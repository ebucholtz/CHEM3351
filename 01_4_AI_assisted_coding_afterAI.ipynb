# Exploring AI-Assisted and “Vibe” Coding in Python

Large language models (LLMs) are trained using massive datasets pulling information from books, websites, and other forms of written communication to learn patterns in language. They essentially are designed to predict the next word in a sentence based on words that come before it. The preditive training that it goes through becomes a method to learn grammar, style, factual information, and some would argue reasoning patterns. One particularly interesting use case is training models on computer code. Code is a computer language with rules (syntax) that must be followed. LLMs can recognize code and common programing structures in the same way they learn to recognize other human languages. The result is they are able to suggest code completion, generate functional code, and potentially explain programming logic. As such, they become powerful tools for individuals learning to code or experienced programmers to general cheminformatics workflows.

## Graphing with Copilot

We haven't learned how to create graphs yet, but the syntax is straight forward. 

The following code cell obtains data from the periodic table available at PubChem. Run the code cell. 

```python
import requests 
import pandas as pd 
url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/periodictable/CSV"
df = pd.read_csv(url)
df.head(5)
```

Open the AI chat with `ctrl` + `alt` + `I`.

Copy the code and paste it into the chat window. Ask the AI to explain the code.

Add a generate code cell below. In the code cell ask the AI to generate a bar graph of ionization energy vs atomic symbol from data in the df table.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,6))
plt.bar(df['Symbol'], df['IonizationEnergy'])
plt.xlabel('Atomic Symbol')
plt.ylabel('Ionization Energy (eV)')
plt.title('Ionization Energy vs Atomic Symbol')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Ask it to redraw the plot again but color the bars by period.

```python
import matplotlib.pyplot as plt

# Get unique periods and assign a color to each
periods = df['ElectronConfiguration'].str.extract(r'\[(\w+)\]')[0].fillna('1')
unique_periods = periods.unique()
colors = plt.cm.tab20.colors
period_color_map = {period: colors[i % len(colors)] for i, period in enumerate(unique_periods)}

# Assign colors to each element based on its period
bar_colors = periods.map(period_color_map)

plt.figure(figsize=(16,6))
plt.bar(df['Symbol'], df['IonizationEnergy'], color=bar_colors)
plt.xlabel('Atomic Symbol')
plt.ylabel('Ionization Energy (eV)')
plt.title('Ionization Energy vs Atomic Symbol (Colored by Period)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
